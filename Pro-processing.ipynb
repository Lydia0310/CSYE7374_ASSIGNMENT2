```python
#integrate into a function to process two original .csv file.
#output will be used on Spark 
def clean_dataset(filepath,file_name):
    columns = ['age',
        'workclass',
        'fnlwgt',
        'education',
        'education-num',
        'marital-status',
        'occupation',
        'relationship',
        'race',
        'sex',
        'capital-gain',
        'capital-loss',
        'hours-per-week',
        'native-country',
        'income']
    df = pd.read_csv(filepath,na_values=' ?',names=columns);
    
    df_cleaned = df.dropna()
    
    categorical_features = [c for c in df_cleaned if not df_cleaned[c].dtype.kind in ('i', 'f')]
    for i in categorical_features:
        df_cleaned[i] = df_cleaned[i].map(lambda x: x.strip(' ')).astype(str)
    cat_df = df_cleaned[categorical_features]
    num_df = df_cleaned.drop(categorical_features,1)
    cat_df_encode = cat_df.apply(lambda x: pd.factorize(x)[0])
    data = pd.concat([num_df, cat_df_encode], axis=1)
    data = data.drop(['fnlwgt','capital-gain','capital-loss',1)
    return data.to_csv(file_name, sep=',',header= False,encoding='utf-8')


```

```python
clean_dataset("adult_data.csv","training_data.csv")
clean_dataset("adult.test.csv","validation_data.csv")
```
